```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
tau_m = 10.0           # Membrane time constant
dt = 1.0               # Time step
V = 0.0                # Initial membrane potential
R = 1.0                # Membrane resistance
V_th = 1.0             # Threshold potential
V_reset = 0.0          # Reset potential
time_steps = 100       # Number of time steps
V_trace = []           # Record the change of membrane potential over time
spike_times = []       # Record spike trigger times

I = np.random.uniform(0.5, 2.5, time_steps)
I = np.round(I, 2)
# Random input current

for t in range(time_steps):

    V = V + (dt / tau_m) * (-V + R * I[t])

    if V >= V_th:
        spike_times.append(t * dt)
        V_trace.append(V)
        V = V_reset
    else:
        V_trace.append(V)
```

```python
plt.figure(figsize=(10, 4))
plt.plot(np.arange(0, time_steps * dt, dt), V_trace, label="Membrane Potential")
plt.axhline(y=V_th, color='r', linestyle='--', label="Threshold")
plt.scatter(spike_times, [V_th] * len(spike_times), color='red', label="Spikes")
plt.xlabel("Time (ms)")
plt.ylabel("Membrane Potential (V)")
plt.title("LIF Neuron Simulation")
plt.legend()
plt.show()
```
